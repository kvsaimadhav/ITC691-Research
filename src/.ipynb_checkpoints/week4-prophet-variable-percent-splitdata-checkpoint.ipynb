# Actual Value vs Predicted Value Check with Prophet Algorithm

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
 
data = pd.read_csv("michigan-history.csv")
data.head()
```

# NewData created from Data for only the needed columns

```python
dimension = data.shape
rows = dimension[0]
cols = dimension[1]
date = []
cases_diff = []
for i in range(0,rows):
    datevalue = data.iloc[i]["date"]
    date.append(datevalue)
    casesvalue = data.iloc[i]["positiveIncrease"]
    cases_diff.append(casesvalue)
records = len(cases_diff)
cases = []
date_reorder = []
for j in range(records-1,-1,-1):
    cases.append(cases_diff[j])
    date_reorder.append(date[j])
newdata = pd.DataFrame(list(zip(date_reorder,cases)))
newdata.columns = ['ds', 'y']
newdata.head()
```

# SplitData with  variation of percent from 0%- 95% of Data for Training Purposes

```python
dimension1 = newdata.shape
rows1 = dimension1[0]
cols1 = dimension1[1]
per = 0.95
splitrows = int(per * rows1)
splitds = []
splity = []
column_names = ["ds", "y"]
for i in range(0,splitrows):
    splitds.append(newdata.iloc[i][0])
    splity.append(newdata.iloc[i][1])
splitdata = pd.DataFrame(list(zip(splitds,splity)),columns = column_names)
splitdata.head()
```

```python
from fbprophet import Prophet

m = Prophet()
m.fit(splitdata)
period = rows - splitrows
future = m.make_future_dataframe(periods = period)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head()
```

```python
fig1 = m.plot(forecast)
fig2 = m.plot_components(forecast)
```

```python
error = []
for i in range(splitrows,rows):
    pred = forecast.iloc[i][1]
    act = newdata.iloc[i][1]
    diff = abs(pred - act)
    error.append(diff)
meanerror = sum(error)/float(period)
print(meanerror)
```

```python
count = 0
for i in range(splitrows,rows):
    predlower = int(forecast.iloc[i][2])
    predupper = int(forecast.iloc[i][3])
    predict =  int(forecast.iloc[i][1])
    actual = int(newdata.iloc[i][1])
    if(predlower > predupper):
        if(actual >= predupper and actual <= predlower):
            count = count + 1
    elif(predlower < predupper):
        if(actual >= predlower and actual >= predupper):
            count = count + 1
    else:
        if(actual == predlower and actual == predupper):
            count = count + 1
print("Accuracy is ", (count/period)*100)
```
