# Usage of Minimum and Saturating Values for Prediction of Data

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
 
data = pd.read_csv("michigan-history.csv")
data.head()
```

```python
dimension = data.shape
rows = dimension[0]
cols = dimension[1]
date = []
cases_diff = []
for i in range(0,rows):
    datevalue = data.iloc[i]["date"]
    date.append(datevalue)
    casesvalue = data.iloc[i]["positiveIncrease"]
    cases_diff.append(casesvalue)
records = len(cases_diff)
cases = []
date_reorder = []
for j in range(records-1,-1,-1):
    cases.append(cases_diff[j])
    date_reorder.append(date[j])
newdata = pd.DataFrame(list(zip(date_reorder,cases)))
newdata.columns = ['ds', 'y']
newdata.head()
```

```python
dimension1 = newdata.shape
rows1 = dimension1[0]
cols1 = dimension1[1]
per = 0.80 # percentage of data to be splitted in seqeuntial order
splitrows = int(per * rows1)
splitds = []
splity = []
column_names = ["ds", "y"]
for i in range(0,splitrows):
    splitds.append(newdata.iloc[i][0])
    splity.append(newdata.iloc[i][1])
splitdata = pd.DataFrame(list(zip(splitds,splity)),columns = column_names)
splitdata.head()
```

```python
from fbprophet import Prophet

m = Prophet()
m.fit(splitdata)
period = rows - splitrows
future = m.make_future_dataframe(periods = period)
future.tail()
```

```python
splitdata["floor"] = 0
future["floor"] = 0
splitdata["cap"] = max(splitdata["y"])
future["cap"] = max(splitdata["y"])
m = Prophet(growth='logistic')
m.fit(splitdata)
fcst = m.predict(future)
fig = m.plot(fcst)
```

```python
fcst[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head()
```

```python
#count = 0
diff = []
for i in range(splitrows,rows):
    #predlower = int(fcst.iloc[i][2])
    #predupper = int(fcst.iloc[i][3])
    predict =  int(fcst.iloc[i][1])
    actual = int(newdata.iloc[i][1])
    diff.append(abs(actual-predict))
    #print(predict, actual, predlower, predupper)
meanerror = sum(diff)/period
print("Mean Error in Prediction for 0-95% data is ", meanerror)
```

```python
count = 0
#diff = []
for i in range(splitrows,rows):
    predlower = int(fcst.iloc[i][2])
    predupper = int(fcst.iloc[i][3])
    predict =  int(fcst.iloc[i][1])
    actual = int(newdata.iloc[i][1])
    #diff.append(abs(actual-predict))
    #print(predict, actual, predlower, predupper)
    if(predlower > predupper):
        if(actual >= predupper and actual <= predlower):
            count = count + 1
    elif(predlower < predupper):
        if(actual >= predlower and actual >= predupper):
            count = count + 1
    else:
        if(actual == predlower and actual == predupper):
            count = count + 1
#meanerror = sum(diff)/period
print("Accuracy in Prediction for 95% data is ", (count/period)*100)
```

```python
splitdata.to_csv('simulation.csv', index = False)
```
